```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
PlayerCount_df = pd.DataFrame({"Total Number of Players":[len(purchase_data)]})
PlayerCount_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
PurchasingAnalysis_df = pd.DataFrame({"Number of Unique Items":[len(pd.unique(purchase_data["Item Name"]))]
                                   ,"Average Price":["{:.2f}".format(purchase_data["Price"].mean())]
                                   ,"Total Number of Purchases": [len(purchase_data)]
                                   ,"Total Revenue":["${:.2f}".format(purchase_data["Price"].sum())]
                                  })

PurchasingAnalysis_df.head()
                                   
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
GenderDemo_df = pd.DataFrame({"Gender" : purchase_data["Gender"]
                              ,"SN": purchase_data["SN"]
                             }) 


RemoveDup = GenderDemo_df.drop_duplicates("SN",inplace = False)
GenderDemo_df = RemoveDup
GenderDemoGroupby_df = GenderDemo_df.groupby(["Gender"])
FinalGenderDemo_df = GenderDemoGroupby_df.count()
FinalGenderDemo_df["Percentage of Players"] =  FinalGenderDemo_df["SN"] /  FinalGenderDemo_df["SN"].sum() *100
FinalGenderDemo_df.columns =  ['Total Count', 'Percentage of Players']
FinalGenderDemo_df['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val ) for val in FinalGenderDemo_df['Percentage of Players']], index = FinalGenderDemo_df.index)
GenderSortFinal_df = FinalGenderDemo_df.sort_values(["Total Count"], ascending=False)
GenderSortFinal_df.rename_axis(None,inplace=True)
GenderSortFinal_df.head()


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#print(TotalDistinctFemale,TotalDistinctMale,TotalDistinctOther )
OnGoingData_df = purchase_data

PurchasingAnalysGroupBy_df = OnGoingData_df.groupby('Gender').agg({'Purchase ID':'count', 'Price':['mean','sum'] })
PurchasingAnalysGroupBy_df.columns = PurchasingAnalysGroupBy_df.columns.droplevel(0)
PurchasingAnalysGroupBy_df.columns = ['Purchase Count','Average Purchase Price','Total Purchase value']

PurchasingAnalysGroupBy_df.loc['Female', 'Avg Total Purchase per Person'] = PurchasingAnalysGroupBy_df.loc['Female', 'Total Purchase value'] / GenderSortFinal_df.loc['Female','Total Count']
PurchasingAnalysGroupBy_df.loc['Male', 'Avg Total Purchase per Person'] = PurchasingAnalysGroupBy_df.loc['Male', 'Total Purchase value'] / GenderSortFinal_df.loc['Male','Total Count']
PurchasingAnalysGroupBy_df.loc['Other / Non-Disclosed', 'Avg Total Purchase per Person'] = PurchasingAnalysGroupBy_df.loc['Other / Non-Disclosed', 'Total Purchase value'] / GenderSortFinal_df.loc['Other / Non-Disclosed','Total Count']


PurchasingAnalysGroupBy_df["Average Purchase Price"] = PurchasingAnalysGroupBy_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
PurchasingAnalysGroupBy_df["Total Purchase value"] = PurchasingAnalysGroupBy_df["Total Purchase value"].astype(float).map("${:,.2f}".format)
PurchasingAnalysGroupBy_df["Avg Total Purchase per Person"] = PurchasingAnalysGroupBy_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)


PurchasingAnalysGroupBy_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
AgeDemographics_df =  purchase_data

# Create bins in which to place values based age
bins = [0, 9, 14, 19, 24,29,34,39,50]
# Create labels for these bins
group_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
# Place the data series into a new column inside of the DataFrame
AgeDemographics_df["Age Group"] = pd.cut(AgeDemographics_df["Age"], bins, labels=group_labels)
#AgeDemographics_df.drop_duplicates("SN",inplace = True)
RemoveDup = AgeDemographics_df.drop_duplicates("SN",inplace = False)
AgeDemographics_df = RemoveDup


AgeDemographicsGroup_df = AgeDemographics_df.groupby(["Age Group"])
FinalAgeDemographicsGroup_df = AgeDemographicsGroup_df[["SN"]].count()

FinalAgeDemographicsGroup_df["Percentage of Players"] = FinalAgeDemographicsGroup_df["SN"]/FinalAgeDemographicsGroup_df["SN"].sum()*100
FinalAgeDemographicsGroup_df = FinalAgeDemographicsGroup_df.rename(columns={'SN': 'Total Count'})
FinalAgeDemographicsGroup_df.rename_axis(None,inplace=True)
FinalAgeDemographicsGroup_df["Percentage of Players"] = FinalAgeDemographicsGroup_df["Percentage of Players"].astype(float).map("{:,.2f}%".format)
FinalAgeDemographicsGroup_df

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PurchaseAnalysisAge_df =  purchase_data

# Create bins in which to place values based age
Pbins = [0, 9, 14, 19, 24,29,34,39,50]
# Create labels for these bins
Pgroup_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
# Place the data series into a new column inside of the DataFrame
PurchaseAnalysisAge_df["Age Group"] = pd.cut(PurchaseAnalysisAge_df["Age"], Pbins, labels=Pgroup_labels)
PurchaseAnalysisAge_df = PurchaseAnalysisAge_df.groupby("Age Group").agg({'Purchase ID':'count', 'Price':['mean','sum']})
PurchaseAnalysisAge_df.columns = PurchaseAnalysisAge_df.columns.droplevel(0)

# Merge the two DataFrames together based on the Dates they share
PurchaseAnalysisAgeFinal_df = pd.merge(PurchaseAnalysisAge_df, FinalAgeDemographicsGroup_df, left_index=True, right_index=True)
PurchaseAnalysisAgeFinal_df = PurchaseAnalysisAgeFinal_df.rename(columns={'count': 'Purchase Count', 'mean':'Average Purchase Price','sum':'Total Purchase Value'})

PurchaseAnalysisAgeFinal_df["Avg Total Purchase per Person by Age Group"] = PurchaseAnalysisAgeFinal_df["Total Purchase Value"] / PurchaseAnalysisAgeFinal_df["Total Count"]
PurchaseAnalysisAgeFinal_df = PurchaseAnalysisAgeFinal_df.drop(columns = ["Total Count","Percentage of Players"])
PurchaseAnalysisAgeFinal_df["Average Purchase Price"] = PurchaseAnalysisAgeFinal_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
PurchaseAnalysisAgeFinal_df["Total Purchase Value"] = PurchaseAnalysisAgeFinal_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
PurchaseAnalysisAgeFinal_df["Avg Total Purchase per Person by Age Group"] = PurchaseAnalysisAgeFinal_df["Avg Total Purchase per Person by Age Group"].astype(float).map("${:,.2f}".format)


PurchaseAnalysisAgeFinal_df
#PurchaseAnalysisAge_df.info()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
TopSpenders_df =  purchase_data
TopSpenders_df = TopSpenders_df.groupby("SN")

TopSpenders_df = pd.DataFrame({"Purchase Count":TopSpenders_df["Purchase ID"].count()
                                      ,"Average Purchase Price":TopSpenders_df["Price"].mean()
                                      ,"Total Purchase Value":TopSpenders_df["Price"].sum()})

TopSpenders_df = TopSpenders_df.sort_values(["Total Purchase Value"], ascending=False)
TopSpenders_df["Average Purchase Price"] = TopSpenders_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
TopSpenders_df["Total Purchase Value"] = TopSpenders_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
TopSpenders_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
MostPopularItem_df =  purchase_data
MostPopularItemGroup_df = MostPopularItem_df.groupby(["Item ID", "Item Name"])

MostPoplarItemCalc_df = pd.DataFrame({"Purchase Count":MostPopularItemGroup_df["Item ID"].count()
                                      ,"Item Price":MostPopularItemGroup_df["Price"].mean()
                                      ,"Total Purchase Value":MostPopularItemGroup_df["Price"].sum()})

MostPopularItemFinal_df = MostPoplarItemCalc_df.sort_values(["Purchase Count"], ascending=False)
MostPopularItemFinal_df["Item Price"] = MostPopularItemFinal_df["Item Price"].astype(float).map("${:,.2f}".format)
MostPopularItemFinal_df["Total Purchase Value"] = MostPopularItemFinal_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
MostPopularItemFinal_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
MostProfitItem_df =  purchase_data
MostProfitItemGroup_df = MostProfitItem_df.groupby(["Item ID", "Item Name"])

MostProfitIteamCalc_df = pd.DataFrame({"Purchase Count":MostProfitItemGroup_df["Item ID"].count()
                                      ,"Item Price":MostProfitItemGroup_df["Price"].mean()
                                      ,"Total Purchase Value":MostProfitItemGroup_df["Price"].sum()})

MostProfitItemFinal_df = MostProfitIteamCalc_df.sort_values(["Total Purchase Value"], ascending=False)
MostProfitItemFinal_df["Item Price"] = MostProfitItemFinal_df["Item Price"].astype(float).map("${:,.2f}".format)
MostProfitItemFinal_df["Total Purchase Value"] = MostProfitItemFinal_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
MostProfitItemFinal_df.head()
```
